```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('train.csv')
fullfillment=pd.read_csv('fulfilment_center_info.csv')
meal=pd.read_csv('meal_info.csv')
```

```python
dataset.head()
```

```python
fullfillment.head()
```

```python
meal.head()
```

```python
dataset1=pd.merge(dataset,fullfillment,on='center_id')
```

```python
final_dataset=pd.merge(dataset1,meal,on='meal_id')
```

```python
final_dataset.drop(['center_id','meal_id'],axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
final_dataset.drop(['id','base_price'],axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
X=final_dataset.drop(['num_orders'],axis=1)
y=final_dataset['num_orders']
```

```python
X=pd.get_dummies(X)
```

```python
X.head()
```

```python
y.value_counts()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=10,random_state=0)
rf.fit(X,y)
```

```python
rf.score(X,y)
```

```python
test_data=pd.read_csv('test_QoiMO9B.csv')
```

```python
test_data.head()
```

```python
test_data1=pd.merge(test_data,fullfillment,on='center_id')
```

```python
full_test_data=pd.merge(test_data1,meal,on='meal_id')
```

```python
full_test_data.head()
```

```python
full_test_data.drop(['center_id','meal_id'],axis=1,inplace=True)
```

```python
full_test_data.head()
```

```python
final_test_data=full_test_data.drop(['id','base_price'],axis=1)
```

```python
final_test_data.head()
```

```python
final_test_data=pd.get_dummies(final_test_data)
```

```python
final_test_data.head()
```

```python
y_predicted=rf.predict(final_test_data)
```

```python
submission = pd.DataFrame({
        "id": test_data['id'],
        "num_orders": y_predicted
    })
```

```python
submission.to_csv('FoodDemandForecasting_submission.csv', index=False)
```
